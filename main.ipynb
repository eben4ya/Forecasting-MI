# **Metode Moving Average**

### Metode Moving Average melibatkan penggunaan data historis untuk mengidentifikasi pola atau tren dengan variabel independentnya adalah waktu.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data harga reksadana obligasi
data = {
    'Date': ['1Feb2024', '1Jan2024', '1Dec2023', '1Nov2023', '1Oct2023', '1Sep2023', '1Aug2023', '1Jul2023', '1Jun2023', '1May2023', '1Apr2023'],
    'Price': [1292, 1286, 1281, 1274, 1281, 1280, 1277, 1271, 1266, 1263, 1260],
    'AUM': [3.81, 3.92, 3.99, 4.67, 4.99, 6.7, 7.23, 7.45, 7.75, 8.58, 9.1]
}

df = pd.DataFrame(data)

# Ubah format tanggal menjadi datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d%b%Y')

# Membuat plot
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price'], marker='o', linestyle='-', label='Price')
plt.plot(df['Date'], df['AUM'], marker='o', linestyle='-', label='AUM')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Price and AUM over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Menghitung moving average dengan window = 3
window = 3
moving_average = df['Price'].rolling(window=window).mean()

# Prediksi harga reksadana pada bulan Maret 2024
price_nov = df.loc[2, 'Price']
price_dec = df.loc[1, 'Price']
price_jan = df.loc[0, 'Price']

moving_average_maret_2024 = (price_nov + price_dec + price_jan) / window

print("Perkiraan harga reksadana pada bulan Maret 2024 dengan metode moving average (window = 3):",
      moving_average_maret_2024)
```

# **Metode Regresi Linear Sederhana**



### Metode Regresi memeriksa hubungan kausal antara satu atau lebih variabel independen (penyebab) dan satu variabel dependen (hasil). Dengan memahami hubungan kausal akan membuat prediksi yang lebih akurat.

# **Implementasi Kode**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data harga reksadana obligasi
data = {
    'Date': ['1Feb2024', '1Jan2024', '1Dec2023', '1Nov2023', '1Oct2023', '1Sep2023', '1Aug2023', '1Jul2023', '1Jun2023', '1May2023', '1Apr2023'],
    'Price': [1292, 1286, 1281, 1274, 1281, 1280, 1277, 1271, 1266, 1263, 1260],
    'AUM': [3.81, 3.92, 3.99, 4.67, 4.99, 6.7, 7.23, 7.45, 7.75, 8.58, 9.1]
}

df = pd.DataFrame(data)

# Ubah format tanggal menjadi datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d%b%Y')

# Membuat plot
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price'], marker='o', linestyle='-', label='Price')
plt.plot(df['Date'], df['AUM'], marker='o', linestyle='-', label='AUM')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Price and AUM over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Memasukkan nilai variabel independen (X) dari pengguna
X_value = float(input("Masukkan nilai variabel independen (X): "))

# Metode Regresi Linear Sederhana
X = df[['AUM']]
y = df['Price']

# Menghitung prediksi harga berdasarkan nilai variabel independen yang dimasukkan
reg = LinearRegression().fit(X, y)
X_new = np.array([[X_value]])
predicted_price = reg.predict(X_new)

print("Koefisien kemiringan (a):", reg.coef_[0])
print("Intercept (b):", reg.intercept_)

print("\nPrediksi menggunakan Metode Regresi Linear Sederhana:")
print("Prediksi harga reksadana obligasi berdasarkan nilai variabel independen (AUM) = ",
      X_value, ":", predicted_price)
```
